```python
import requests
import pandas as pd
from api_keys import weather_key, nps_key
import json

# Replace YOUR_API_KEY with your actual API key
# url = 'https://developer.nps.gov/api/v1/parks?api_key=YOUR_API_KEY'
url_parks = "https://developer.nps.gov/api/v1/parks"
params = {
    'api_key': nps_key
#     'limit': '50'
}
# Send a GET request to the API and retrieve the response
# response = requests.get(url_parks, params=params)

# Convert the response to a JSON object
# data_parks = response.json()

#data_parks
```

```python
parks = []
while True:
    response = requests.get(url_parks, params=params)
    data = response.json()
    park_data = data['data']
    parks.extend(park_data)
    if len(parks) >= int(data['total']):
        break
    params['start'] = str(len(parks))  # Set the start parameter to retrieve the next batch of parks
```

```python
parks_df = pd.DataFrame(parks)
parks_df.head()
```
